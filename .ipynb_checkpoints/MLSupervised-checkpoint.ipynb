# Supervised Learning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline


import warnings
warnings.filterwarnings('ignore')
```

# Column Rename

```python
data.rename(columns={'bp': 'blood_pressure', 'sg': 'specific gravity','al':'albumin','su':'sugar','rbc':'red blood cells','pc':'pus cell','pcc':'pus cell clumps','ba':'bacteria','bgr':'blood glucose random','bu':'blood urea','sc':'serum creatinine','sod':'sodium','pot':'potassium','hemo':'hemoglobin','pcv':'packed cell volume','wc':'white blood cell count','rc':'red blood cell count','htn':'hypertension','dm':'diabetes mellitus','cad':'coronary artery disease','appet':'appetite','pe':'pedal edema','ane':'anemia','classification':'class'},inplace=True)
     
```

# Changing categorical to numerical

```python
# Label enconder

from sklearn.preprocessing import LabelEncoder as Lb

encoder = Lb() #create an object of the Label Encoder class
```

```python
data['red blood cells'] = encoder.fit_transform(data['red blood cells'])
data['pus cell'] = encoder.fit_transform(data['pus cell'])
data['pus cell clumps'] = encoder.fit_transform(data['pus cell clumps'])
data['bacteria'] = encoder.fit_transform(data['bacteria'])
data['hypertension'] = encoder.fit_transform(data['hypertension'])
data['diabetes mellitus'] = encoder.fit_transform(data['diabetes mellitus'])
data['coronary artery disease'] = encoder.fit_transform(data['coronary artery disease'])
data['coronary artery disease'] = encoder.fit_transform(data['coronary artery disease'])
data['appetite'] = encoder.fit_transform(data['appetite'])
data['pedal edema'] = encoder.fit_transform(data['pedal edema'])
data['anemia'] = encoder.fit_transform(data['anemia'])
data['class'] = encoder.fit_transform(data['class'])





```

```python
 #Dropping all columns with incomplete record to avoid inaccuracy
data2 = data.dropna()
print(data2.shape)
```

```python
data2
```

```python
data2.isnull().sum()
```

```python
data2.drop('id', axis = 1, inplace =True)
```

```python
data2
```

```python
data2.columns
```

```python
plt.figure(figsize = (20, 10))

sb.heatmap(data2.corr(), annot = True, linewidths = 2, linecolor = 'b')
plt.show()
     
```

```python
# [['specific gravity', 'albumin', 'blood glucose random', 'blood urea', 'serum creatinine', 'sodium', 'hemoglobin', 'hypertension', 'diabetes mellitus']]
```

```python
# feature extraction or engineering
x =  data2[['specific gravity', 'albumin', 'blood glucose random', 'blood urea', 'serum creatinine', 'sodium', 'hemoglobin', 'hypertension', 'diabetes mellitus']]
# x = data2[['age', 'blood_pressure', 'hypertension', 'diabetes mellitus', 'coronary artery disease', 'anemia', 'hemoglobin' ]]
y = data2['class'].values




from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

model = LogisticRegression()

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=20)

model.fit(x_train, y_train)
```

```python
x.corr()
```

```python
model.predict(x_test)
```

```python
print(f'Accuracy = {model.score(x_train, y_train)}')
```

```python
print(f'Accuracy = {model.score(x_test, y_test)}')
```

# Confusion Matrix

```python
# confusion matrix: this gives a matrix for the predicted and actual
y_pred = model.predict(x_test)

confusion_matrix = pd.crosstab(y_test, y_pred, rownames= ['Actual'], colnames = ['Predicted'])
sb.heatmap(confusion_matrix, annot=True)
```

# KNN

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score


model = KNeighborsClassifier(1)

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state=20)

model.fit(x_train, y_train)

```

```python
 y_pred = model.predict(x_test)
```

```python
print(f'The accuracy of the model is {accuracy_score(model.predict(x_train), y_train)}')
```

```python
print(f'The accuracy of the model is {accuracy_score(y_pred, y_test)}')
```

# Optimal K

```python
scores = []
for k in range(1,20):
  model = KNeighborsClassifier(k)
  model.fit(x_train, y_train)
  y_pred = model.predict(x_test)
  print(f'{k} : {accuracy_score(y_pred,y_test)}')
  scores.append(accuracy_score(y_pred, y_test))


# k = 7 is the optimal k because that is the best score k can get. 
# 
```

# Decision Tree

```python
#importing relevant libraries
from sklearn.model_selection import train_test_split
from sklearn import tree   
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
#splitting data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3,random_state=20)
```

```python
#applying tree algorithm
model = DecisionTreeClassifier(min_samples_leaf =12)


model.fit(x_train, y_train)   #fitting our model

y_pred=model.predict(x_test)   # evaluating our model


print("score:{}".format(accuracy_score(y_test, y_pred)))
```

```python
accuracy_score(model.predict(x_train), y_train)
```

```python
import graphviz
graph_ = tree.export_graphviz(model,label='all', rounded=True, filled=True)
graph = graphviz.Source(graph_)

graph.render('data')
graph
```

# Random Forest

```python
#importing relevant libraries
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

#splitting data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=0)


model = RandomForestClassifier(n_estimators=5)
model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_train)

print(f'The accuracy for the train data {metrics.accuracy_score(y_train,y_train)}')
```

```python
y_pred=model.predict(x_test)   # evaluating our model


print(f'The accuracy for the test data {metrics.accuracy_score(y_pred,y_test)}')
```

```python
# FEATURE ENGINEERING IS VERY IMPORTANT.

from sklearn.model_selection import cross_val_score

model = RandomForestClassifier(n_estimators=5)
scores = cross_val_score(model,x,y, cv=5)

print(f'Accuracy on each fold: {scores}')
print(f'Accuracy on each accuracy: {sum(scores)/5}')
```

# Combined Plot

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure(figsize=(16,8))

# Define the models and their accuracy scores
models = ['Logistic Regression', 'KNN', 'Decision Tree', 'Random Forest']
accuracy_scores = [1.0, 0.9, 0.95, 1.0]

# Specify the colors for each bar
colors = ['grey', 'blue', 'purple', 'orange']


# Plot the bar graph
x_pos = np.arange(len(models))
plt.bar(x_pos, accuracy_scores, align='center' , color=colors)
plt.xticks(x_pos, models)
plt.ylabel('Accuracy')
plt.xlabel('Models')
plt.title('Accuracy Comparison of Models')
plt.show()
```
